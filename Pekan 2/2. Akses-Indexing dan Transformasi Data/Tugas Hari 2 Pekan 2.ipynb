<h2>Soal 2: Filter DataFrame Dengan Boolean</h2>

1. Tampilkan semua data siswa yang level pendidikan orang tuanya 'high school' dan 'bachelor degree'
2. Tampilkan semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'
3. Tampilkan semua data siswa yang mempunyai nilai lebih dari 70, berjenis kelamin perempuan dengan test preparation course 'completed'

```python
# 2.1 semua data siswa yang level pendidikan orang tuanya 'high school' dan 'bachelor degree'
#code here

pendidikan1 = df['parental level of education'] == 'high school'
pendidikan2 = df['parental level of education'] == 'bachelor degree'

df[(pendidikan1) | (pendidikan2)]
```

```python
# 2.2 semua data siswa yang mempunyai nilai lebih dari 50 dengan test preparation course 'completed'

#code
df.loc[(df['nilai'] > 50) & (df['test preparation course'] == 'completed')]
```

```python
# 2.3 semua data siswa yang mempunyai nilai lebih dari 70, berjenis kelamin perempuan dengan test preparation course 'completed'

#code
df.loc[(df['nilai'] > 70) & (df['gender'] == 'female') & (df['test preparation course'] == 'completed')]
```

<h2>Soal 3: Data Transformation</h2>

lakukan data transformasi terhadap column 'nilai' dengan fungsi transformasi, berikut rumus dari normalisasi

![alt text](https://drive.google.com/uc?id=1wed-EbRSTA0JP1ceDp6lcKBqH2yJ4aa3)

```python
#code here
def normalisasi(x):
    return (x - (x.min()/(x.max()-x.min())))

print(df.nilai)
print('nilai minimum: ',df.nilai.min())
print('nilai maximum: ',df.nilai.max())
print('nilai min/(max-min): ',df.nilai.min()/(df.nilai.max()-df.nilai.min()))
print('\nHasil Normalisasi Nilai:')
df[['nilai']].apply(normalisasi)
```
