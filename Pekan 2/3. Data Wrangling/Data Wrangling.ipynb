**Data Wrangling**

Data wrangling adalah proses dimana kita mengambil data asli kemudian mengubahnya menjadi sebuah format yang mudah dianalisis. Dalam praktiknya, terdapat 3 proses yang umum dilakukan dalam proses wrangling:



*   Data cleaning
*   Data transformation
*   Data enrichment


**Data Cleaning**

Data cleaning adalah proses yang perlu dilakukan ketika kita membaca data set. Beberapa proses data cleaning yang penting adalah:


*   Renaming
*   Sorting & reordering
*   Data type conversions
*   Deduplicating data
*   Addresing missing or invalid data
*   filtering to the desired subset of data
 
Untuk poin missing data akan dibahas pada pertemuan ke 4 dan filtering sudah dibahas pada pertemuan sebelumnya.

```python
import pandas as pd

df_kebakaran = pd.read_csv('drive/My Drive/Bootcamp Data Science Python/Pekan 2/3. Data Wrangling/dataset_18.csv', sep=';')
df_kebakaran
```

Kita bisa melakukan data wrangling untuk memudahkan kita analisis data. Jika kita lihat pada kolom, terdapat kolom Benda dan nilai. untuk memudahkan analisis, kita bisa mengubah label kolom tersebut dengan sebuah nama yang lebih mudah dimengerti. Perubahan nama kolom, bisa menggunakan atribut .rename():

```python
df_kebakaran.rename(columns={'Benda': 'Benda yang terbakar',
                   'nilai': 'kerugian (rupiah)'},
          inplace=True)

#inplace = True digunakan supaya kita tidak perlu menyimpan lagi ke dalam sebuah variabel hasil dari perubahan yang kita lakukan.
df_kebakaran.head()
```

```python
# Kita juga bisa melakukan perubahan pada kolom dengan mengubahnya menjadi uppercase seperti berikut:
df_kebakaran.rename(str.upper, axis='columns').columns
```

Tahapan selanjutnya dalam data wrangling adalah, kita perlu melakukan perubahan terhadap tipe data kolom jika tipe tersebut tidak sesuai. untuk mengecek setiap tipe data kita gunakan syntaks seperti berikut:


```python
df_kebakaran.dtypes
```

Jika kita lihat, semua tipe data berupa objek. kita perlu melakukan perubahan terhadap tipe data berupa datetime atau numerik. Untuk mengubah tipe data, kita menggunakan atribut .astype() seperti berikut:

```python
df_kebakaran.astype({'Tanggal':'datetime64[ns]'})
```
