**Soal 4. Time grouping**

 

*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
#code here
climate['2016'].groupby(pd.Grouper(freq='1M'))[['meantemp','humidity','wind_speed','meanpressure']].max()
```

Expected output:

(Gambar hanya sebagian)

![alt text](https://drive.google.com/uc?id=1s2Y8Dc3M3xbG-51mTuTmXUFgLvHDvc5Q)

```python
# Buat dataframe baru untuk rata-rata climate per 1 bulan pada 2016

rata_rata = climate['2016'].groupby(pd.Grouper(freq='1M'))[['meantemp','humidity','wind_speed','meanpressure']].mean().copy()
rata_rata
```

```python
# Rata-rata Meantemp tertinggi:
max_meantemp = rata_rata['meantemp'].max()

#print(max_meantemp)
rata_rata.loc[rata_rata['meantemp'] == max_meantemp]
```

```python
# Rata-rata Humidity Tertinggi
max_humidity = rata_rata['humidity'].max()

rata_rata.loc[rata_rata['humidity'] == max_humidity]
```

```python
# Rata-rata Wind speed tertinggi

max_wind_speed = rata_rata['wind_speed'].max()

rata_rata.loc[rata_rata['wind_speed'] == max_wind_speed]
```

```python
# Rata-rata Meanpressure tertinggi

max_meanpressure = rata_rata['meanpressure'].max()

rata_rata.loc[rata_rata['meanpressure'] == max_meanpressure]
```

Jawab disini:


1.   Meantemp tertinggi terjadi pada bulan: Juni ***2016-06-30***
2.   Humidity tertinggi terjadi pada bulan: Januari ***2016-01-31***
3.   Wind speed tertinggi terjadi pada bulan: April ***2016-04-30***
4.   Meanpressure tertinggi terjadi pada bulan: Maret ***2016-03-31***
