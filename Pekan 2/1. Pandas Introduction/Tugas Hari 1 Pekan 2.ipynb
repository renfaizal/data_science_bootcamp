<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 

**Perbedaan Series dengan list dan dictionary:**

Series mempunyai index yang dapat difungsikan sebagai key, sedangkan pada list hanya memiliki index dan pada dictionary hanya memiliki key. Serta index pada series dapat disesuaikan dengan keinginan/kebutuhan kita.

**Perbedaan dataframe dan series:**
*   Series adalah objek 1 dimensi
*   DataFrame adalah suatu objek 2 dimensi, mirip seperti tabel

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!

Jawab: 

Deskriptif statistik digunakan untuk menggambarkan/summary data. Seperti : nilai mean, median, modus, variance, standar deviasi, quartile, range dan sejenisnya. Descriptive statistics berlaku ketika pengumpulan seluruh sample data (populasi) dapat dilakukan.

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here

print(df_.describe())
print('\n',df_['price'].mean())
print('\n',df_['depth'].std())
print('\n',df_['carat'].max())
```
