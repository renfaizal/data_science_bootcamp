<h2>Soal 3: Preparing Data</h2>

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
import numpy as np
df_region_sales_trans = df_region_sales.transpose()
```

```python
fig,ax = plt.subplots(figsize=(16,8))
x = np.arange(4)
width = 0.07
color = ['r','#348abd','#988dd4','#777777','#fac05d','#8db941','#feb5b8','#e24a32','#348abd','#978ed4','#777777','#fac05d']

for i,j in enumerate(df_region_sales_trans.columns):

  ax.bar(x+(width*i),df_region_sales_trans[j], width=width, label = j, color = color[i], align='edge')

ax.set_title('\nMean Sales Video Games by Genre', fontdict={'weight':'bold', 'size':15})
ax.set_ylabel('Mean Sales', fontdict={'size':13})
ax.set_xlabel('Region Sales', fontdict={'size':13})

ax.legend(bbox_to_anchor=(1.05, 1.12),frameon= True)
plt.xticks(x+0.4, labels = df_region_sales_trans.index)
plt.style.use('seaborn')
plt.show()
```

Insight yang dapat diperoleh dari visualisasi di atas diantaranya:


*   Visualisasi bar chart menggambarkan rata-rata penjualan video games di beberapa region menurut genre game.
*   Rata-rata penjualan video games tertinggi terdapat di North America (NA_Sales) dan rata-rata penjualan terendah berada di Other_Sales
*   Rata-rata penjualan games bergenre platform merupakan yang tertinggi di NA_Sales, sedangkan game bergenre shooter merupakan yg terendah di NA_Sales.
