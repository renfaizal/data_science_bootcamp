```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = [random.random()*100 for i in range(100)]

plt.figure(figsize=(12, 8)) # Membuat figure/kotak untuk plot
plt.plot(x, y)
plt.show()
```

***Figure***, adalah window atau page atau halaman dalam objek visual. kalau kita ngegambar di kertas, maka kertas tersebutlah yang di namakan figure.

***Axes***, kedalam Figure kita menambahkan Axes, Axes adalah suatu area di dalam figure dimana data akan di plot. Dalam Axes juga terdapat berbagai macam komponen lagi seperti x-axis, y-axis, dan lain sebagainya. hal ini akan kalian lihat di artikel ke dua.

```python
'''
Cara Eksplisit
'''

# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# membuat data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

# menambahkan dalam ke objek axes
ax.plot(data_x, data_y)

# menampilkan data
plt.show()
```

```python
'''
Cara Implisit
'''
# membuat objek visual
plt.plot(data_x, data_y)

# menampilkan objek visual
plt.show()
```

```python
# membuat objek figure dan axes
fig, ax = plt.subplots()

# membuat data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

# menambahkan dalam ke objek axes
ax.plot(data_x, data_y)

# menampilkan data
plt.show()
```

```python
fig, ax = plt.subplots()
data_x1 = [1, 2, 3, 4]
data_y1 = [10, 20, 25, 30]
data_x2 = [1, 2, 3, 4, 5]
data_y2 = [5, 15, 20, 25, 30]
ax.plot(data_x1, data_y1)
ax.plot(data_x2, data_y2)
plt.show()
```
