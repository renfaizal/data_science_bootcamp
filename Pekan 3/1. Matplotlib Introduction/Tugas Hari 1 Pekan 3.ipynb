**Figure**, adalah window atau page atau halaman dalam objek visual.

**Axes**, kedalam Figure kita menambahkan Axes, Axes adalah suatu area di dalam figure dimana data akan di plot. Dalam Axes juga terdapat berbagai macam komponen lagi seperti x-axis, y-axis, dan lain sebagainya.

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)

# Figure dan Axis kosong tanpa Data
fig = plt.figure()
ax = fig.add_subplot()

# menampilkan objek visual
plt.show()
```

```python
# membuat figure dan axis secara explicit
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x, y)
plt.show()
```

```python
# membuat figure dan axis secara implisit
plt.figure()
plt.plot(x, y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
from IPython.display import Image

fig, ax = plt.subplots()
# Plot ketiga data berikut ke dalam satu axis
ax.plot(x, y, x, y2, x, y3)
# gunakan seaborn style
plt.style.use('seaborn')
plt.show()

# save data tersebut dengan nama file bebas
fig.savefig('outputNo3.png')

# tampilkan kembali gambar
Image('outputNo3.png')
```
