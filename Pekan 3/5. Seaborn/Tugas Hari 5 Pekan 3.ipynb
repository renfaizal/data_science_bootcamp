```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df.head()
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
 
#code here

clarity_ranking = ["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"]

sns.scatterplot(data=df, x="carat", y="price", hue="clarity", size="depth", palette=palette, hue_order=clarity_ranking)
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
sns.set_theme(style="white", palette=None)

clarity_ranking = ["IF", "VVS1", "VVS2", "VS1", "VS2", "SI1", "SI2", "I1"]

g = sns.catplot(x="clarity", y="price", col="cut", data=df, kind="box", showfliers = False)

g.set_axis_labels("", "Price")
g.set_xticklabels(clarity_ranking)
g.set_titles("{col_name} {col_var}")
g.despine(left=True)
```

Expected Output:

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)
sns.set_theme()
#code here

markers=['p', 'o', 's', 'D', 'X'] 
sns.pairplot(df,x_vars=['carat', 'depth', 'table', 'price'],y_vars=['carat','depth', 'table', 'price'], hue='cut', markers=markers, corner=True)
```
