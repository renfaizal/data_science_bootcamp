<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here

fig, ax = plt.subplots(figsize=(18,8), nrows=1, ncols=3, sharey = True)

ax[0].plot(x, tan, label='tanX', color='b')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, label='sinX', color='r')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, label='cosX', color='g')
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax = plt.subplots(figsize=(14,8))

ax.plot(x, y)
ax.set_title('Grafik CosX')

# upper left
ins1 = ax.inset_axes([-0.09, 0.9, 0.27, 0.3])
ins1.plot(x,y5,color='y')
ins1.plot(x,y,color='b')
ins1.set_title('Grafik CosX^5')

# upper right
ins1 = ax.inset_axes([0.91, 0.9, 0.27, 0.3])
ins1.plot(x,y4,color='black')
ins1.plot(x,y,color='b')
ins1.set_title('Grafik CosX^4')

# bottom left
ins1 = ax.inset_axes([-0.09, -0.1, 0.27, 0.3])
ins1.plot(x,y2,color='r')
ins1.plot(x,y,color='b')
ins1.set_title('Grafik CosX^2')

# bottom right
ins1 = ax.inset_axes([0.91, -0.1, 0.27, 0.3])
ins1.plot(x,y2,color='g')
ins1.plot(x,y,color='b')
ins1.set_title('Grafik CosX^3')


plt.show()
```
